##**4 Feature engineering**

Feature engineering é o termo utilizado para definir um conjunto de técnicas utilizado na criação e manipulação de features (recursos), tendo como objetivo desenvolver um bom modelo de aprendizado de máquina. A engenharia de recursos é o processo de usar o conhecimento de domínio dos dados
para criar recursos (vetores de recursos) que fazem os algoritmos de aprendizado de máquina funcionarem.

vetor de características é um vetor n-dimensional de características numéricas que representam algum objeto.
Muitos algoritmos em aprendizado de máquina requerem uma representação numérica de objetos,
uma vez que tais representações facilitam o processamento e a análise estatística.

```python
X, y = train_data.values, target.values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.1, random_state=0)
```

Os métodos ensemble funcionam melhor quando os preditores são tão independentes uns dos outros quanto possível. Um
maneira de obter diversos classificadores é treiná-los usando algoritmos muito diferentes. Isso aumenta a chance
que eles cometerão tipos de erros muito diferentes, melhorando a precisão do conjunto.

Classificação com Random Forest, Naive Bayes e SVC

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import VotingClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
GNB= GaussianNB()
dt = DecisionTreeClassifier( criterion='gini',
                                    splitter='best',
                                    max_depth=4,
                                    max_features=1)
rnd_clf = RandomForestClassifier(n_estimators=250, criterion='entropy', max_features=1, max_depth=350, min_samples_split=3)
svm_clf = SVC(kernel='rbf', C = 1)
voting_clf = VotingClassifier(
estimators=[('svc', svm_clf), ('GNB', GNB), ('rf', rnd_clf), ('dt', dt)],
voting='hard')
voting_clf.fit(X_train, y_train)
```

Resultado dos classificadores individuais e do classificador por votação

```python
from sklearn.metrics import accuracy_score
for clf in (GNB, svm_clf, rnd_clf, dt, voting_clf):
  clf.fit(X_train, y_train)
  y_pred = clf.predict(X_test)
  print('A acurácia em percentual é: ', clf.__class__.__name__, round(accuracy_score(y_test, y_pred)*100,2), '%')
```

Matriz de confusão

```python
from sklearn import metrics
```

```python
metrics.ConfusionMatrixDisplay.from_predictions(y_test, y_pred, cmap='Blues');
plt.grid(False)
```
